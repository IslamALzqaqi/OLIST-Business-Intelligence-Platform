```python
orders = datasets["olist_orders_dataset"]

date_columns = [
    "order_purchase_timestamp",
    "order_approved_at",
    "order_delivered_carrier_date",
    "order_delivered_customer_date",
    "order_estimated_delivery_date"
]

for col in date_columns:
    orders[col] = pd.to_datetime(orders[col])

datasets["olist_orders_dataset"] = orders
```

```python
orders[date_columns].dtypes
```

```python
orders = datasets["olist_orders_dataset"].copy()

orders["order_year"] = orders["order_purchase_timestamp"].dt.year
orders["order_quarter"] = orders["order_purchase_timestamp"].dt.quarter
orders["order_month"] = orders["order_purchase_timestamp"].dt.month
orders["order_month_name"] = orders["order_purchase_timestamp"].dt.month_name()
orders["order_day"] = orders["order_purchase_timestamp"].dt.day
orders["order_day_name"] = orders["order_purchase_timestamp"].dt.day_name()
orders["order_week"] = orders["order_purchase_timestamp"].dt.isocalendar().week.astype("int")
orders["order_weekday"] = orders["order_purchase_timestamp"].dt.weekday

datasets["olist_orders_dataset"] = orders
```

```python
orders.head()
```

```python
orders = datasets["olist_orders_dataset"].copy()

orders["delivery_days"] = (
    orders["order_delivered_customer_date"] -
    orders["order_purchase_timestamp"]
).dt.days

datasets["olist_orders_dataset"] = orders
```

```python
orders[[
    "order_purchase_timestamp",
    "order_delivered_customer_date",
    "delivery_days"
]].head(10)
```

```python
orders["delay_days"] = (
    orders["order_delivered_customer_date"] -
    orders["order_estimated_delivery_date"]
).dt.days
```

```python
orders["delivery_status"] = orders["delay_days"].apply(
    lambda x: "Early" if x < 0
    else "On Time" if x == 0
    else "Late"
)
```

```python
order_items = datasets["olist_order_items_dataset"].copy()
orders = datasets["olist_orders_dataset"].copy()

fact_sales = order_items.merge(
    orders,
    on="order_id",
    how="left"
)

fact_sales.head()
```

```python
payments = datasets["olist_order_payments_dataset"].copy()

payment_summary = (
    payments
    .groupby("order_id", as_index=False)
    .agg(
        total_payment=("payment_value", "sum"),
        max_installments=("payment_installments", "max"),
        payment_count=("payment_sequential", "count")
    )
)
```

```python
dim_customers = fact_sales[
    [
        "customer_id",
        "customer_unique_id",
        "customer_zip_code_prefix",
        "customer_city",
        "customer_state"
    ]
].drop_duplicates()

print(dim_customers.shape)
dim_customers.head()
```

```python
dim_products = fact_sales[
    [
        "product_id",
        "product_category_name",
        "product_name_lenght",
        "product_description_lenght",
        "product_photos_qty",
        "product_weight_g",
        "product_length_cm",
        "product_height_cm",
        "product_width_cm"
    ]
].drop_duplicates()

print(dim_products.shape)
```

```python
dim_sellers = fact_sales[
    [
        "seller_id",
        "seller_zip_code_prefix",
        "seller_city",
        "seller_state"
    ]
].drop_duplicates()

print(dim_sellers.shape)
```

```python
start_date = fact_sales["order_purchase_timestamp"].min().normalize()
end_date = fact_sales["order_purchase_timestamp"].max().normalize()

dim_date = pd.DataFrame({
    "Date": pd.date_range(start=start_date, end=end_date, freq="D")
})
```

```python
dim_date["Year"] = dim_date["Date"].dt.year
dim_date["Quarter"] = dim_date["Date"].dt.quarter
dim_date["Quarter_Name"] = "Q" + dim_date["Quarter"].astype(str)

dim_date["Month"] = dim_date["Date"].dt.month
dim_date["Month_Name"] = dim_date["Date"].dt.month_name()

dim_date["Day"] = dim_date["Date"].dt.day
dim_date["Day_Name"] = dim_date["Date"].dt.day_name()

dim_date["Week"] = dim_date["Date"].dt.isocalendar().week.astype(int)

dim_date["Weekday"] = dim_date["Date"].dt.weekday + 1

dim_date["Is_Weekend"] = dim_date["Weekday"].isin([6, 7])

dim_date["Year_Month"] = dim_date["Date"].dt.strftime("%Y-%m")
```

```python
category_columns = [
    "order_id",
    "customer_id",
    "product_id",
    "seller_id",
    "order_status",
    "delivery_status"
]

for col in category_columns:
    fact_sales_final[col] = fact_sales_final[col].astype("category")
```

```python
import os

os.makedirs("clean_data", exist_ok=True)
```

```python
fact_sales_final.to_csv("clean_data/Fact_Sales.csv", index=False)

dim_customers.to_csv("clean_data/Dim_Customers.csv", index=False)

dim_products.to_csv("clean_data/Dim_Products.csv", index=False)

dim_sellers.to_csv("clean_data/Dim_Sellers.csv", index=False)

dim_date.to_csv("clean_data/Dim_Date.csv", index=False)
```
